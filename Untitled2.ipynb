```python
import pandas as pd
import numpy as np

from sklearn.metrics.pairwise import cosine_similarity
```

```python
df_movies  = pd.read_csv("./Datasets/Movies.csv")
df_links   = pd.read_csv("./Datasets/Links.csv")
df_tags    = pd.read_csv("./Datasets/Tags.csv")
df_ratings = pd.read_csv("./Datasets/Ratings.csv")
```

```python
user_item_matrix = df_ratings.pivot_table(index='userId', columns='movieId', values='rating').fillna(0)
user_item_matrix.head()
```

```python
item_similarity = cosine_similarity(user_item_matrix.T)

item_similarity_df = pd.DataFrame(item_similarity, 
                                  index = user_item_matrix.columns, 
                                  columns=user_item_matrix.columns)

item_similarity_df
```

```python
item_similarity_df.mean(axis=1)
```

```python
item_similarity_df.min(axis=1)
```

```python
item_similarity_df.max(axis=1)
```

```python
def recommend_movies ( user_id,
                      user_item_matrix,
                      item_similarity_df,
                      top_n = 10
                     ):
    rated_movies = user_item_matrix.loc[user_id]
    rated_movies = rated_movies[rated_movies > 0 ].index.tolist()
    
    sim_scores = item_similarity_df[rated_movies].sum(axis=1)
    
    sim_scores = sim_scores.drop(index=rated_movies)
    recommend_movie_ids = sim_scores.sort_values(ascending = False).head(top_n).index
    recommend_movies = df_movies[df_movies["movieId"].isin(recommend_movie_ids)]
    
    return recommend_movies
```

```python
user_id_sample = 2
```

```python
recommend_movies_sample = recommend_movies(user_id_sample, user_item_matrix, item_similarity_df, top_n = 5)
recommend_movies_sample
```
